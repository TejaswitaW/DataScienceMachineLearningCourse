```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating pandas dataframe
original_dataframe=pd.read_csv("IMDB-Movie-Data.csv")
```

```python
# converting all column names into lower case
original_dataframe.columns=[col.lower() for col in original_dataframe]
```

```python
rating=original_dataframe["rating"].unique()
len(rating)
```

```python
# getting type of rating
type(rating)
```

```python
rating_sum=original_dataframe.groupby(["rating"])["rating"].sum()
rating_sum.shape
```

```python
# we need to convert x-axis data i.e rating into integer
# for better understanding
rating=rating.astype('int32')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(rating,rating_sum)
plt.title("Movie Rating With Its Sum")
plt.xlabel("Movie Rating")
plt.ylabel("Rating Sum")
plt.show()
```

```python
# ratings_sum is series
# plotting pie plot for rating_sum
rating_sum.plot.pie()
```
