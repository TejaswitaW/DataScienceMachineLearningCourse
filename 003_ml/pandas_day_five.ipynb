### Data Visualization

```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plotting simple plot of data points
plt.plot([1,2,3],[4,5,6])
```

```python
# plotting sin wave
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
```

```python
# plt.figure() create a new figure 
fig=plt.figure()
# add an axes to the figure
# [0,0,1,1] means length,bottom,width,height
# for 2D plot we make width=1 and height=1
ax=fig.add_axes([0,0,1,1])
ax.set_title("sine wave")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.plot(x,y)
```

```python
# creating simple bar plot language vs student count
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
langs=["C","C++","Java","Python","PHP"]
students=[23,17,35,29,19]
ax.bar(langs,students)
plt.show()
```

```python
data=([[30,25,50,20],[40,23,51,17],[35,22,45,19]])
X=np.arange(4)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
# data[0]=[30,25,50,20]
# data[1]=[40,23,51,17]
# data[2]=[35,22,45,19]
ax.bar(X+0.00,data[0],color="b",width=0.25)
ax.bar(X+0.25,data[1],color="g",width=0.25)
ax.bar(X+0.50,data[2],color="r",width=0.25)
```

```python
# we can take different data points and plot them
X=original_dataframe.rating
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
#ax=fig.add_axes([0,0,1,1])
ax.bar(X+0.00,original_dataframe["revenue (millions)"],color="b",width=0.25)
# "votes" value is very high ,so it is divided by 2000
ax.bar(X+0.25,original_dataframe["votes"]/2000,color="g",width=0.25)
ax.bar(X+0.50,original_dataframe["runtime (minutes)"],color="r",width=0.25)
```

```python
fig,ax=plt.subplots(1,1)
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
# bins is range,gives count of data,data partition,gives data partition count
# example how many values between 0 and 100
ax.hist(original_dataframe["revenue (millions)"],bins=[0,100,200,300,400,500])
ax.set_title("histogram of movies")
ax.set_xticks([0,100,200,300,400,500])
ax.set_xlabel("revenue")
ax.set_ylabel("no.of movies")
plt.show()
```

```python
#pichart
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis("equal")
langs=["C","C++","Java","Python","PHP"]
students=[23,17,35,29,19]
ax.pie(students,labels=langs,autopct="%1.2f%%")
plt.show()
#take count of movie and plot in pichart rating
```

```python
# plotting scatter plot
girls_grades=[89,90,89,100,80,90,100,80,34]
boys_grades=[30,29,49,100,48,38,45,20,30]
grade_range=[20,30,40,50,60,70,80,90,100]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(grade_range,girls_grades,color="r")
ax.scatter(grade_range,boys_grades,color="b")
ax.set_xlabel("Grades Range")
ax.set_ylabel("Grades Scored")
ax.set_title("Scatter Plot")
plt.show()
```
